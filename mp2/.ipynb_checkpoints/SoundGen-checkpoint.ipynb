```python
# Simple sine
samples = fs*duration

t = np.linspace(0,duration,samples)

f = fref*np.power(2, note/12)
w = 2*np.pi*f

x = 0.2*np.sin(w*t) + 0.1*np.sin(5/4*w*t) + 0.1*np.sin(3/2*w*t)
```

```python
class Note:

    # Duration is in bars
    def  __init__(self, number, duration, intensity):
        self.number = number
        self.duration = duration
        self.intensity = intensity
        if not ((len(self.number) == len(self.duration)
                 and len(self.number) == len(self.intensity))):
            raise ValueError('Number, duration, and intensity list sizes must match.')

        
    def __repr__(self):
        return self.__str__()
    
    def __str__(self):
        return  "note\t\t" + str(self.number)
    
    def summary(self):
        return  ("note\t\t" + str(self.number) +
                 "\nduration\t" + str(self.duration) +
                 "\nintenisty\t" + str(self.intensity))
    
    def count(self):
        return len(self.number)
```
